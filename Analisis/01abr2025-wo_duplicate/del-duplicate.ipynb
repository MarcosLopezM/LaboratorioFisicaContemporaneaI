<h1> Eliminando archivos duplicados </h1>

```python
import os
import shutil
import filecmp
```

```python
"""
    Número de mediciones para:
    - Fondo sin plasma
    - Plasma sin abrir shutter
    - Plasma con shutter
    - Fondo apagado
"""
num_mediciones = [10, 68, 299, 13]
```

```python
def move_unique_files(src_path, dest_path):
    if not os.path.exists(dest_path):
        os.makedirs(dest_path)

    archivos = sorted(os.listdir(src_path))
    archivos_copiados = {}  # Diccionario para guardar archivos únicos

    for archivo in archivos:
        src_file = os.path.join(src_path, archivo)
        
        # Verifica si el archivo es único comparándolo con los ya copiados
        es_duplicado = False
        for unique_file in archivos_copiados:
            if filecmp.cmp(src_file, archivos_copiados[unique_file], shallow=False):
                es_duplicado = True
                break
        
        if not es_duplicado:
            dest_file = os.path.join(dest_path, archivo)
            shutil.copy2(src_file, dest_file)
            archivos_copiados[archivo] = src_file
            print(f"{archivo} movido.")
```

```python
src_path_UV = "../01abr2025/UV"
dest_path_UV = "../01abr2025-wo_duplicate/UV"
src_path_VIS = "../01abr2025/VIS"
dest_path_VIS = "../01abr2025-wo_duplicate/VIS"
```

<h2> Archivos de la carpeta UV </h2>

```python
move_unique_files(src_path_UV, dest_path_UV)
```

<h2> Archivos de la carpeta VIS </h2>

```python
move_unique_files(src_path_VIS, dest_path_VIS)
```
